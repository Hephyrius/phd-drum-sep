```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data preparation
data = {
    'Model': ['Hybrid Demucs', 'Hybrid Demucs', 'Hybrid Demucs', 'Hybrid Demucs', 'Hybrid Demucs',
              'HT Demucs', 'HT Demucs', 'HT Demucs', 'HT Demucs', 'HT Demucs',
              'Conv-TasNet', 'Conv-TasNet', 'Conv-TasNet', 'Conv-TasNet', 'Conv-TasNet'],
    'Drum': ['Bass', 'Snare', 'Hi-Hat', 'Tom', 'Cymbal + Ride',
             'Bass', 'Snare', 'Hi-Hat', 'Tom', 'Cymbal + Ride',
             'Bass', 'Snare', 'Hi-Hat', 'Tom', 'Cymbal + Ride'],
    'F1 Measure': [0.785, 0.728, 0.789, 0.963, 0.919,
                   0.782, 0.719, 0.769, 0.937, 0.927,
                   0.784, 0.665, 0.746, 0.914, 0.887],
    'Jaccard Index': [0.701, 0.626, 0.723, 0.933, 0.863,
                      0.695, 0.612, 0.690, 0.888, 0.876,
                      0.695, 0.537, 0.656, 0.854, 0.817]
}

df = pd.DataFrame(data)

# Create separate plots for F1 Measure and Jaccard Index
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Define bar width and positions
bar_width = 0.25
indices = np.arange(len(df['Drum'].unique()))
model_names = df['Model'].unique()

# Plot F1 Measure
for i, model in enumerate(model_names):
    values = df[df['Model'] == model]['F1 Measure']
    position = indices + bar_width * i
    axes[0].bar(position, values, bar_width, label=model)

axes[0].set_xlabel('Drum Types')
axes[0].set_ylabel('F1 Measure')
axes[0].set_title('F1 Measure Comparison of Models by Drum Types')
axes[0].set_xticks(indices + bar_width)
axes[0].set_xticklabels(df['Drum'].unique())
axes[0].legend()
axes[0].grid(True)

# Plot Jaccard Index
for i, model in enumerate(model_names):
    values = df[df['Model'] == model]['Jaccard Index']
    position = indices + bar_width * i
    axes[1].bar(position, values, bar_width, label=model)

axes[1].set_xlabel('Drum Types')
axes[1].set_ylabel('Jaccard Index')
axes[1].set_title('Jaccard Index Comparison of Models by Drum Types')
axes[1].set_xticks(indices + bar_width)
axes[1].set_xticklabels(df['Drum'].unique())
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()

# Show plot
plt.show()
```
