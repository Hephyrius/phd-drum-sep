# Imports

```python
import museval
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
accuracy = [1.0, 0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1, 0]
sdrs = []
length_s = "4s"
for t in accuracy:
    df = pd.read_csv(f"D:/Github/phd-drum-sep/rebuttal_experiments/conv_tasnet_model_results_epoch_200_accuracy_{t}_drop//results.csv")
    results = museval.EvalStore(frames_agg='median', tracks_agg='median')
    results.df = df
    sdrs.append(results.agg_frames_tracks_scores()['drums']['SDR'])


```

```python
# Set up the seaborn style
sns.set(style="whitegrid")

# Plotting with seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x=accuracy, y=sdrs, marker='o', color='b')
plt.xlabel('Accuracy')
plt.ylabel('SDR')
plt.title('SDR vs. Accuracy')
plt.grid(True)
plt.show()
```
